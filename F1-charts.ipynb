```python
from matplotlib import pyplot as plt
import fastf1
import fastf1.plotting

fastf1.plotting.setup_mpl()

session = fastf1.get_session(2024, 'Australia', 'R')
session.load()
```

```python
def driver(name1, name2, color1, color2, label1, label2):
    laps1 = session.laps.pick_driver(name1)
    laps2 = session.laps.pick_driver(name2)
    plt.plot(laps1.LapNumber, laps1.LapTime, c=color1, label=label1)
    plt.plot(laps2.LapNumber, laps2.LapTime, c=color2, label=label2)
    plt.legend()
    plt.show()
```

```python
driver('LEC', 'NOR', 'green', 'orange', 'Leclerc', 'Norris')
```
